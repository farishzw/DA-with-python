### 数据框选择
- 获取数据getting: 
    - 列: 类似R，用方框选取列名
    - 行: 切片，用方框截取
- 通过标签筛选:df.loc
    - 行:方框后第一个参数，用具体列中的数据或者列的相对索引
    - 列:列名或者含有列名的列表
- 通过相对位置整数筛选:df.iloc
    - 行:方框后第一个参数
    - 列:切片
    - 框:切片和列表
- 根据某一列的取值筛选
    - df2[df2["E"].isin(["two", "four"])]

```python
#选择核心数据进行数据分析，根据相对位置的数字筛选列
dfc = df.iloc[:,[11,12,19,25,33,42,49,57,58,59,60,61,63,65,67,69,71]]#三大维度：政务服务满意度、办事体验满意度、群众办事便利度
```

```python
#重命名列名
dfc.columns = ['窗口','办事效率','办事流程','业务技能','工作纪律','投诉处理','政务公开','服务态度','服务环境','咨询导办','窗口设置','四减落实','跨域通办','十项服务','一件事一次办','在线办理','自助服务']
#根据列的值筛选
dfc[dfc['窗口'].isin(['E.市审批局','C.市社保局'])]
```

### 数据预处理
- 将输入时的object类型转换为数值类型
    - pd.to_numeric,astype('Int64'):并分别调用apply()函数和pd.to_numeric()函数将该列的字符串数据转换为数值型数据，同时将无法转换的数据设置为缺失值（NaN）。最后，使用astype()函数将该列的数据类型转换为int类型，同时将NaN值转换为可空整数（Int64）类型。
- 字符串替换可计算的数值
    - 问卷中不涉及的指标通常替换为Nan:调用replace函数匹配字符串，使用正则表达式
    - 将所有以"不清楚"开头的字符串替换为空值dfr = dfc.replace(r'^F.不清楚.*$', np.nan, regex=True),其中np.nan代表缺失值
- 数据去重（以下待补充）
- 数据缺失值处理等
- 数据标准化等

```python
#将dataframe中的object对象转为int64
columns_to_convert = ['投诉处理', '四减落实', '跨域通办','十项服务','一件事一次办','在线办理','自助服务']
for col in columns_to_convert:
    dfc[col] = dfc[col].apply(pd.to_numeric, errors='coerce').astype('Int64')
```

```python
#将窗口中的字段规范化
dfr = dfc.replace('[A-Z\.]',"",regex=True)
```

### 数据数据操作

- stats求均值
    - 按行求均值:df.mean(1)
    - 按列求均值:df.mean()
    - 按组求均值: groupby()【】.mean(),groupby传入参数，单个或多个以列表形式表示，方框参数为待求均值的字段，求和、方差等类似改变函数
- 处理具有不同维度的数据时自动对齐，为赋值的用空值填补；此外panda在特定的维度自动broadcast
- apply:将定义的函数作用在数据上
    - df.apply(np.cumsum)
    - df.apply(lambda x:x.max() - x.min())作用列X
- histogramming: 计算频率s.value_counts()
- 字符串处理:series中配备了一系列方法在str属性中，如lower()等

```python
# 计算每个部门的平均得分，三大维度细分指标的平均值
avg_scores = dfr.groupby("窗口").mean()
print('部门平均得分：\n',avg_scores)
```

```python
#自动补齐
#创建index
dates = pd.date_range("20130101",periods=6)
#唯一一个数值，第一个数值不要，从第二个数字开始用
pd.Series([1, 3, 5, np.nan, 6, 8], index=dates).shift(1)
#df.sub(s,axis="index")
```

### 数据高级操作
- 增加数据
    - 增加列:df['new_column'] = values
    - 增加行:
        - 使用loc方法：df.loc[len(df)] = values
        - 使用字典：df = df.append({"col1": value1, "col2": value2}, ignore_index=True)
- 数据合并: merge:pd.merge(df1,df2,on='key')
    - pd.concat():数据按行聚合，直接合并
    - pd.merge按列聚合

```python
com_df = pd.merge(avg_scores,df_minus,on='窗口')
pd.DataFrame(range(0,21),com_df.columns)
```

```python
df1 = com_df.reindex(columns=['窗口']+ ['总得分']+list(com_df.columns[1:7]) + ["政务满意度得分"] +list(com_df.columns[7:11])+['办事体验满意度']+list(com_df.columns[11:17])+['群众办事便利度']+list(com_df.columns[17:21])+['政务服务供给能力'])
pd.DataFrame(range(0,26),df1.columns)
```

```python
df1.loc[:,'政务满意度得分'] = df1.iloc[:,2:8].mean(1)*6
df1.loc[:,'办事体验满意度'] = df1.iloc[:,9:13].mean(1)*4
df1.loc[:,'群众办事便利度'] = df1.iloc[:,14:20].mean(1)*6
df1.loc[:,'政务服务供给能力'] = 20-df1.iloc[:,21:25].mean(1)*4
df1.loc[:,'总得分'] = df1.loc[:,'政务满意度得分'] + df1.loc[:,'办事体验满意度'] + df1.loc[:,'群众办事便利度'] + df1.loc[:,'政务服务供给能力']
```

```python
df1 = df1.sort_values(by='总得分',ascending=False)
df1
```
